# Define dispersion relation

```python
def omega(k):
    return 1.0 * k
```

```python
k_array = np.linspace(-2,2,100)

plt.plot(k_array, omega(k_array))
```

# Define plane wave

```python
def wave(x, t, k):
    phase = k * x - omega(k) * t
    wave = np.exp(1j * phase)
    return wave
```

# Animate!

### Define spatial grid

```python
x = np.linspace(-5,5,100)
```

### Define time points at which each frame will be drawn

```python
time_array = np.linspace(0,5,100)
```

### Define data to be drawn for each frame

```python
k1 = 1.0
def get_frame_data(i):
    t = time_array[i]
    return wave(x,t,k1).real
```

### Animate

```python
fig = plt.figure()
ax = fig.gca()

ax.set_xlim(-5,5)
ax.set_ylim(-1,1)

line, = ax.plot([],[])

def update_line_data(frame_index):
    y = get_frame_data(frame_index)
    line.set_data(x, y)
    return (line,)


from matplotlib.animation import FuncAnimation

ani = FuncAnimation(fig, func=update_line_data)
```
